```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
from dateutil.parser import parse
import numpy as np

import warnings
import datetime as dt
warnings.filterwarnings('ignore')

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
events = pd.read_csv('data/events_up_to_01062018.csv')
events.head()
```

```python
events.count()
```

# Analisis TP1

```python
events.columns
```

```python
events.drop(['url','skus','city','search_engine','campaign_source','channel'], inplace=True, axis=1)
```

```python
#Cambiamos el tipo de dato de fecha de timestamp a datetime para un mejor analisis y renombramos a "fecha"
events[['timestamp']] = events[['timestamp']].apply(pd.to_datetime)
events.rename({
    'timestamp' : 'fecha',
}, axis=1, inplace=True)
events.head()
```

```python
#Detalles de Productos que fueron comprados
productos = events.loc[:, ['event','fecha','person', 'sku', 'model', 'condition', 'storage', 'color']]
productos_comprados = productos.loc[productos.event == 'conversion']
productos_comprados.drop(['event'], inplace=True, axis=1)
productos_comprados.head()
```

```python
productos_comprados.count()
```

```python
# productos_comprados = productos_comprados.drop_duplicates(['person','sku'])
productos_comprados = productos_comprados.drop_duplicates()
productos_comprados.count()
```

```python
#Detalles de Productos que fueron visitados
# productos_visitados = productos.loc[((productos.event == 'viewed product') | (productos.event == 'checkout'))]
productos_visitados = productos.loc[productos.event == 'viewed product']
productos_visitados.drop(['event'], inplace=True, axis=1)
productos_visitados.head()
```

```python
productos_visitados.count()
```

```python
# productos_visitados = productos_visitados.drop_duplicates(['person','sku'])
productos_visitados = productos_visitados.drop_duplicates()
productos_visitados.count()
```

```python
# Eliminamos de productos visitados las personas que estan en productos comprados
productos_visitados = productos_visitados.merge(productos_comprados, on='person', how='left')
productos_visitados = productos_visitados[productos_visitados['fecha_y'].isnull()]
productos_visitados = productos_visitados.loc[:,['fecha_x','person', 'sku_x', 'model_x', 'condition_x', 'storage_x', 'color_x']]
productos_visitados.rename(columns={'fecha_x': 'fecha', 'sku_x': 'sku', 'model_x': 'model', 'condition_x': 'condition', 'storage_x': 'storage', 'color_x': 'color'}, inplace=True)
productos_visitados.head()
```

```python
productos_visitados.count()
```

```python
#Caracterisitcas del dispositivos con el que el usuario visito el sitio
visitas = events.loc[:,['event','fecha','person','new_vs_returning','region','country','device_type','screen_resolution','operating_system_version','browser_version']]
features_usuario = visitas.loc[visitas.event == 'visited site']
features_usuario.drop(['event'], inplace=True, axis=1)
features_usuario.head()
```

```python
features_usuario.count()
```

```python
# features_usuario = features_usuario.drop_duplicates(['person','new_vs_returning','region','country','device_type','screen_resolution','operating_system_version','browser_version'])
features_usuario = features_usuario.drop_duplicates()
features_usuario.count()
```

```python
del events
del productos
del visitas
```

# Set de Datos de Entrenamiento

```python
user_prod_comprados = productos_comprados.merge(features_usuario, on="person", how="inner")
user_prod_comprados.head()
```

```python
user_prod_comprados.count()
```

```python
user_prod_vistas = productos_visitados.merge(features_usuario, on="person", how="inner")
user_prod_vistas.head()
```

```python
user_prod_vistas.count()
```

```python
del features_usuario
del productos_comprados
del productos_visitados
```

```python
#concateno
user_prod_comprados['conversion'] = 1
user_prod_vistas['conversion'] = 0
frames = [user_prod_comprados, user_prod_vistas]
sets = pd.concat(frames)
sets.head()
```

```python
del frames
del user_prod_comprados
del user_prod_vistas
```

```python
sets.count()
```

```python
# Rellenamos los screen_resolution que estan en nan con resolucion 0x0
sets = sets.fillna(value='0x0')
```

## Categorizacion de los features a entrenar

```python
# creo una nueva columna person_int para entrenar transformando el valor alfanumerico de person a numerico
sets['person_int'] = labelencoder.fit_transform(sets['person'])
sets.head()
```

```python
# convertimos a integer el valor flotante de sku
sets['sku'] = sets['sku'].astype('int')
sets.head()
```

```python
#Categorizamos los valores de los features con label encoder
sets['model'] = labelencoder.fit_transform(sets['model'])
sets['condition'] = labelencoder.fit_transform(sets['condition'])
sets['storage'] = labelencoder.fit_transform(sets['storage'])
sets['color'] = labelencoder.fit_transform(sets['color'])
sets['new_vs_returning'] = labelencoder.fit_transform(sets['new_vs_returning'])
sets['region'] = labelencoder.fit_transform(sets['region'])
sets['country'] = labelencoder.fit_transform(sets['country'])
sets['device_type'] = labelencoder.fit_transform(sets['device_type'])
sets['screen_resolution'] = labelencoder.fit_transform(sets['screen_resolution'])
sets['operating_system_version'] = labelencoder.fit_transform(sets['operating_system_version'])
sets['browser_version'] = labelencoder.fit_transform(sets['browser_version'])
sets.head()
```

```python
#Categorizamos la fecha del visto del producto en dia hora y minuto (ya que el anio es 2018)
sets['dia_view_prod'] = sets['fecha_x'].apply(lambda x: x.day).astype('int')
sets['mes_view_prod'] = sets['fecha_x'].apply(lambda x: x.month).astype('int')
sets['dia_sem_view_prod'] = sets['fecha_x'].apply(lambda x: x.dayofweek).astype('int')
sets = sets.drop('fecha_x',1)
```

```python
#Categorizamos la fecha de visita del usuario en dia mes hora y minuto (ya que el anio es 2018)
sets['dia_visit_user'] = sets['fecha_y'].apply(lambda x: x.day).astype('int')
sets['mes_visit_user'] = sets['fecha_y'].apply(lambda x: x.month).astype('int')
sets['dia_sem_visit_user'] = sets['fecha_y'].apply(lambda x: x.dayofweek).astype('int')
sets = sets.drop('fecha_y',1)
```
